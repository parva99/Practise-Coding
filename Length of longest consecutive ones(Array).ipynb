```python
#Question
# Given a binary string A. It is allowed to do at most one swap between any 0 and 1. Find and return the length of the longest consecutive 1â€™s that can be achieved.


class Solution:
    # @param A : string
    # @return an integer
    
    def solve(self, A):
    
        def maximum_one(arr,n):
            max_one=0
            maxi=0
            temp=0
    
            for i in range(0,n):
                if(arr[i]==1):
                    max_one=max_one+1
                    temp=temp+1
                else:
                    maxi=max(maxi,temp)
                    temp=0
            maxi=max(maxi,temp)
    
            left=[0]*n
            right=[0]*n
    
            left[0]=arr[0]
            right[n-1]=arr[n-1]
    
            for i in range(1,n):
                if(arr[i]==1):
                    left[i]=left[i-1]+1
                else:
                    left[i]=0
    
            for i in range(n-2,-1,-1):
                if(arr[i]==1):
                    right[i]=right[i+1]+1
                else:
                    right[i]=0
      
            
            
            sum=0
            for i in range(1,n-1):
                if(arr[i]==0):
                    sum=left[i-1]+right[i+1]
                    if (sum<max_one):
                        maxi=max(maxi,sum+1)
                    else:
                        maxi=max(maxi,sum)
            return maxi
        
    
    
        n=len(A)
        arr=[0]*n
        k=0
        for i in A:
            arr[k]=int(i)
            k=k+1
    
    
        ans=maximum_one(arr,n)
        return ans


```
